```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("photocompounds_database_crude.csv")
```

```python
## Uniformisation des majuscules/minuscules pour une comparaison plus facile

data['Compound Category'] = data['Compound Category'].str.title()
data['Name'] = data['Name'].str.capitalize()
data['Synonym'] = data['Synonym'].str.title()
data['Solvent'] = data['Solvent'].str.capitalize()

## Suppression des doublons
data.drop_duplicates(inplace=True)


## Suppression des colonnes non désirées
data.drop(data.index[0], axis=0, inplace=True)
data.drop(data.columns[5], axis=1, inplace=True)
data.drop(data.columns[6], axis=1, inplace=True)
data.drop(data.columns[6], axis=1, inplace=True)

# Supprimer la partie de l'URL spécifique dans la colonne "Compound Category"
data['Compound Category'] = data['Compound Category'].str.replace('Https://Photochemcad.Com/Databases/Common-Compounds/', '', regex=False)

# Remplacer les tirets par des espaces
data['Compound Category'] = data['Compound Category'].str.replace('-', ' ')

print(data.head())
```

```python
print("    Informations générales :")
print(data.info())
print("    Cinq premières lignes :")
print(data.head())
print("    Nom des colonnes :")
print(data.columns)
print("    Type de données des colonnes :")
print(data.dtypes)
print("    Valeurs manquantes par colonne :")
print(data.isnull().sum()) 
print("    Nombre de doublons :")
print(data.duplicated().sum())
print("    Doublons dans les données :")
print(data[data.duplicated()])
print("    Résumé statistique :")
print(data.describe())

## Vérification des valeurs manquantes
print("\n Résumé après nettoyage :")
print(data.info())
print("\n Cinq premières lignes après nettoyage :")
print(data.head())
```

```python
## Enregistrement du fichier nettoyé
data.to_csv("photocompounds_database.csv", index=False)
print("\n Données nettoyées et sauvegardées dans 'photocompounds_database.csv'")
```
